# Creating local agents with OLLAMA & smolagents

## 1. Smoke Test

```python
from smolagents import LiteLLMModel

model = LiteLLMModel(
    model_id="ollama_chat/qwen2:7b",     # Any Ollama-supported chat model
    api_base="http://127.0.0.1:11434",   # Ollama’s OpenAI-compatible endpoint
    num_ctx=4096,                        # Adjust per model capability
)

messages = [
  {"role": "user", "content": [{"type": "text", "text": "Hello, how are you?"}]}
]

print(model(messages))
```

## 2. Creating the 'Web Search' tool

```python
from smolagents import LiteLLMModel, Tool, CodeAgent
from ddgs import DDGS

class WebSearchTool(Tool):
    name = "web_search"
    description = "Search the web and return the top result snippet."
    inputs = {"query": {"type": "string", "description": "search query"}}
    output_type = "string"

    def forward(self, query: str) -> str:
        res = DDGS().text("python programming", max_results=5)
        string_res = ''.join([str(d) for d in res])
        return string_res if string_res else "No result."

agent = CodeAgent(
    tools=[WebSearchTool()],
    model=model,
    max_steps=4,
    name="ResearchAgent",
)

print(agent.run("Find one reason why AI agents are useful for automating workflows. Cite briefly."))
```
